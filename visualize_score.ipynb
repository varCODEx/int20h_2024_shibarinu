```python
import cv2
import mediapipe
import numpy as np
import pandas as pd
import glob
import os
import tqdm
import matplotlib.pyplot as plt
```

```python
differences = pd.read_csv('img_enface_score.csv')
differences
```

```python
plt.figure(figsize=(16,12))

for i in range(9):
    img = cv2.imread(differences.loc[i, 'img_id'])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(3,3,i+1)
    plt.imshow(img)
```

```python
plt.figure(figsize=(16,12))

for i in range(9):
    img = cv2.imread(differences.loc[len(differences)-1-i-9, 'img_id'])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(3,3,i+1)
    plt.imshow(img)
```
